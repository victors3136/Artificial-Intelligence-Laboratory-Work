```python
def spinner(pdf: list[float]) -> int:
    '''
    Conside an unfair probability spinner with n slots. Each slot has a probabilty P_n ∈[0,1] and ∑_(i=0)^(n−1) P_i =1.
    Write a function that random generate an index i with the probability P_i from the set {0,1,...,n}.
    '''
    if not all(0<= p <= 1 for p in pdf):
        raise ValueError("This is not a valid PDF -- probabilities must belong to [0,1]")
    if not np.isclose(np.sum(pdf),1):
        raise ValueError("This is not a valid PDF -- probabilities must sum up to 1")
    # cdf = [np.sum(pdf[:index + 1]) for index in range(len(pdf))]
    random_nr = np.random.random()
    # for index, c_distr in enumerate(cdf):
    #    if random_nr <= c_distr:
    #        return index 
    # return len(pdf) - 1
    cdf = 0
    for index, prob in enumerate(pdf):
        cdf += prob
        if random_nr <= cdf:
            return index
    return len(pdf) - 1
```

```python
def angle_spinner(angle_list: list[float], type='') -> int:
    match(type):
        case 'deg':
            max = 360
        case 'rad':
            max = 2 * np.pi
        case '': # consider normal pdf
            max = 1
        case _: 
            raise ValueError("Unaccepted type {0}. Accepted types are 'deg', 'rad', ''.".format(type))
    circle = np.sum(angle_list)
    if circle != max:
        raise ValueError("That's not a circle")
    normalised = [angle/max for angle in angle_list]
    return spinner(normalised)
```

```python
np.random.seed(5032024)
def montecarlo_I(n: int) -> float:
    '''
    Consider a large pipe with a radius R=30cm that has the opening pointing straight up, as in the figure. 
    Around this pipe is a rectangular wall with an edge l=1m.
    Balls of radius r=1cm fall down uniformly from above inside the square yard.
    Compute, using a Monte Carlo simulation, the probability that a ball falls inside the pipe. 
    Compare the results with the mathematically deduced probability
    '''
    x_pipe_center = 50
    y_pipe_center = 50 # the probability should not change so long as the pipe is entirely within the box
    counter = 0
    for index in range(n):
        x = np.random.random() * 100
        y = np.random.random() * 100
        if np.sqrt((x - x_pipe_center)**2 + (y - y_pipe_center) ** 2) <= 30:
            counter += 1
    return counter / n

def montecarlo_II(iter_count: int, pipe_center_x: int = 50, pipe_center_y: int = 50, pipe_radius: int = 30, box_len: int = 100)-> float:
    counter = 0
    for index in range(iter_count):
        x = np.random.random() * box_len
        y = np.random.random() * box_len
        if np.sqrt((x - pipe_center_x) ** 2 + (y - pipe_center_y) ** 2) <= pipe_radius:
            counter += 1
    return counter / iter_count

def montecarlo_III(iter_count: int, pipe_center_x: int = 50, pipe_center_y: int = 50, pipe_radius: int = 30, box_len: int = 100)-> float:
    return np.sum([1 
                 if np.sqrt((np.random.random() * box_len - pipe_center_x) ** 2 + (np.random.random() * box_len - pipe_center_y) ** 2) <= pipe_radius
                else 0
                 for index in range(iter_count)
                ]) / iter_count
```

```python
box_area = 100 ** 2# cm^2
pipe_area = np.pi * (30 ** 2) # cm ^2
geometric_prob = pipe_area/box_area
print("|------------------------------------------------------------------|")
print("| {:^10} | {:^15} | {:^15} | {:^15} |".format("Iterations", "Monte Carlo I" , "Monte Carlo II", "Monte Carlo III"))
print("|------------|-----------------|-----------------|-----------------|")
for index in range(2, 7):
    iterations = 10 ** index
    result_I = montecarlo_I(iterations)
    result_II = montecarlo_II(iterations)
    result_III = montecarlo_III(iterations)
    print("| {:>10} | {:<15} | {:<15} | {:<15} |".format(iterations, result_I, result_II, result_III))
print("|------------|-----------------------------------------------------|")
print("| {:^10} | {:^51} |".format("Geo Prob", geometric_prob))
print("|------------------------------------------------------------------|")
```
